# PLANTILLA PARA MODELOS DE REGRESIÓN

En este punto del código se importan las librerías que se quieren usar en el desrrollo del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

En este punto se importa el dataset con el que se va a trabajar con este modelo, provienen de un csv, y tras ejecutar el código, podemos ver cómo almacena pandas el dataset.

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

Una vez cargado el dataset, almacenamos las variables dependientes en una nueva variable llamada X.

```python
X = dataset.iloc[:, 1:2].values                                                                 # Indicamos que queremos desde la primera columna (:) hasta la última, salvo la última (:-1) !!!!!![filas, columnas]!!!!
X
```

Ahora almacenamos la variable dependiente en una variable llamada y.

```python
y = dataset.iloc[:, 2].values                                                                  # Indicamos que solo queremos la última columna <-- VARIABLE A PREDECIRy
y
```

Sección de ajuste del modelo de random forest

```python
from sklearn.ensemble import RandomForestRegressor                                               # Librería del modelo 
regression = RandomForestRegressor(n_estimators=10000, random_state=0)                           # Creación de instancia del modelo
regression.fit(X, y)                                                                             # Se ajusta el modelo
```

En este bloque se utiliza el modelo creado para predecir futuros valores

```python
regression.predict([[6.5]])                                                                     # Se puede observar el aumento de la precision del modelo
```

Finalmente, con el modelo y los datos, generamos gráficas para mostrar la evolución del modelo de manera analítica

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regression.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.ticklabel_format(style='plain', axis='y')                                                       # Para poder mostrar los valores reales que hay en el eje de las y
plt.show()
```
